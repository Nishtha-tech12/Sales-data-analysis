```python
#Import pandas Library for data manipulation(like excel tables)
import pandas as pd
#Import matplotlib for creating basic graphs 
import matplotlib.pyplot as plt
#Import seaborn for advanced and good-looking visualizations
import seaborn as sns
# Set style for cleaner plots
sns.set_style("whitegrid")
```

```python
# Load the dataset from the data folder into a DataFrame
df = pd.read_csv("../data/train.csv")
# Show first 5 rows to understand how data looks
df.head()
```

```python
# Fill missing values in Postal Code with 0
df["Postal Code"] = df["Postal Code"].fillna(0)
```

```python
# Show all column names
list(df.columns)
```

```python
# Group data by Category and calculate total sales
category_sales = df.groupby("Category")["Sales"].sum()
```

```python
# Display all the values
category_sales
```

```python
# Display the category which has highest sales
category_sales.idxmax()
```

```python
# Plot bar chart for category-wise sales
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

### Insight 1:
The Technology category contributes the highest total sales among all categories.

```python
# Group data by Sub-Category and calculate total sales
subcat_sales = df.groupby("Sub-Category")["Sales"].sum()
```

```python
# Display all values
subcat_sales
```

```python
#Display the Sub category in which there are highest sales
subcat_sales.idxmax()
```

```python
# Plot bar chart for sub-category sales
subcat_sales.plot(kind="bar", figsize=(10,5))

plt.title("Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.show()
```

### Insight 2:
The Phones sub-category has the highest sales compared to other sub-categories.
